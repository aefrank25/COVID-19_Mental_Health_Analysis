```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go


df1['Value'] = pd.to_numeric(df1['Value'], errors='coerce')

df1 = df1.dropna(subset=['Value'])

us_states = ["Alabama", "Alaska", "Arizona", "Arkansas", "California", "Colorado", "Connecticut", "Delaware",
             "Florida", "Georgia", "Hawaii", "Idaho", "Illinois", "Indiana", "Iowa", "Kansas", "Kentucky",
             "Louisiana", "Maine", "Maryland", "Massachusetts", "Michigan", "Minnesota", "Mississippi",
             "Missouri", "Montana", "Nebraska", "Nevada", "New Hampshire", "New Jersey", "New Mexico",
             "New York", "North Carolina", "North Dakota", "Ohio", "Oklahoma", "Oregon", "Pennsylvania",
             "Rhode Island", "South Carolina", "South Dakota", "Tennessee", "Texas", "Utah", "Vermont",
             "Virginia", "Washington", "West Virginia", "Wisconsin", "Wyoming"]

filtered_df1 = df1[df1['State'].isin(us_states)]

states = filtered_df1.groupby(['Indicator', 'State'])['Value'].mean().reset_index()


state_anxiety_disorder = states[states['Indicator'] == 'Symptoms of Anxiety Disorder']

xvalue = state_anxiety_disorder['State']
yvalue = state_anxiety_disorder['Value']

fig = go.Figure(data=[go.Bar(x=xvalue, y=yvalue)])

fig.update_layout(
    title={'text': 'Prevalence of Reported Anxiety Symptoms by State', 'x': 0.5, 'xanchor': 'center', 'yanchor': 'top', 'font': {'size': 18}},
    xaxis_title={'text': 'State', 'font': {'size': 14}},
    yaxis_title={'text': 'Prevalence (% of Total Population)', 'font': {'size': 14}},
    xaxis_tickangle=-90)

fig.show()
```

### **Top 10 States and their respective indicators**

```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)

states = df1.groupby(['Indicator', 'State'])['Value'].mean().reset_index()

pivot_table = states.pivot_table(index='State', columns='Indicator', values='Value')

pivot_table = pivot_table.reset_index()

pivot_table['Average'] = pivot_table.select_dtypes(include=['float', 'int']).mean(axis=1)

top_10 = pivot_table.sort_values(by='Average', ascending=False).head(10)

top_10
```

```python
import pandas as pd
import plotly.graph_objects as go

states = df1.groupby(['Indicator', 'State'])['Value'].mean().reset_index()

pivot_table = states.pivot_table(index='State', columns='Indicator', values='Value').reset_index()

pivot_table['Average'] = pivot_table.select_dtypes(include=['float', 'int']).mean(axis=1)

top_10 = pivot_table.sort_values(by='Average', ascending=False).head(10)

fig = go.Figure()

for indicator in states['Indicator'].unique():
    fig.add_trace(go.Bar(
        x=top_10['State'],
        y=top_10[indicator],
        name=indicator
    ))

fig.update_layout(
    barmode='group',
    title={'text': 'Top 10 States by Mental Health Condition Prevalence', 'x': 0.5, 'xanchor': 'center', 'yanchor': 'top', 'font': {'size': 18}},
    xaxis_title={'text': 'State', 'font': {'size': 14}},
    yaxis_title={'text': 'Average Value', 'font': {'size': 14}},
    xaxis_tickangle=-90,
    legend_title_text='Indicator'
)

fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

# Group and filter data
states = df1.groupby(['Indicator', 'State'])['Value'].mean().reset_index()
arizona = states[states['State'] == 'Arizona'].copy()

# Calculate percentages
total_value = arizona['Value'].sum()
arizona['Percentage'] = (arizona['Value'] / total_value) * 100

# Adjust percentages to sum to exactly 100%
arizona['Percentage'] = arizona['Percentage'].round(1)  # Round to one decimal place
difference = 100 - arizona['Percentage'].sum()
arizona.loc[arizona['Percentage'].idxmax(), 'Percentage'] += difference  # Add the difference to the largest slice

# Define colors
colors = ['#0072B2', '#D55E00', '#009E73']

# Create the pie chart
fig = go.Figure(go.Pie(
    labels=arizona['Indicator'],
    values=arizona['Percentage'],
    marker=dict(colors=colors),
    hoverinfo='label+percent',
    textinfo='label+percent',
    textposition='outside',
    pull=[0.1, 0, 0],
))

# Update layout
fig.update_layout(
    title={
        'text': 'Distribution of Mental Health Symptoms in Arizona',
        'x': 0.5,
        'xanchor': 'center',
        'yanchor': 'top',
        'font': {'size': 18}
    },
    height=400,
    legend=dict(
        x=1,
        y=0,
        xanchor='left',
        yanchor='bottom'
    )
)

# Show the chart
fig.show()
```

## **Question 2**

```python
df2 = pd.read_excel('CIS 591_ Indicators of Anxiety and Depression Data.xlsx', sheet_name='Gender Identity')
df2.head()

gender_data = df2[df2['Subgroup'].isin(['Cis-gender female', 'Cis-gender male', 'Transgender'])]
gender_stats = gender_data.groupby('Subgroup')['Value'].agg(['mean', 'max', 'min']).T

styled_gender_stats = (
    gender_stats.style
    .set_table_styles({
        ('index',): [{'selector': 'th', 'props': 'font-weight: bold;'}],
        ('',): [{'selector': 'th.row_heading', 'props': 'font-weight: bold;'}]
    })
    .format("{:.2f}")
)

styled_gender_stats
```

```python
import pandas as pd
import plotly.graph_objects as go

gender_data = df2[df2['Subgroup'].isin(['Cis-gender female', 'Cis-gender male', 'Transgender'])]

gender_stats = gender_data.groupby('Subgroup')['Value'].agg(['mean', 'max', 'min']).T

fig = go.Figure()

for stat in gender_stats.index:
    fig.add_trace(go.Bar(
        x=gender_stats.columns,
        y=gender_stats.loc[stat],
        name=stat
    ))

fig.update_layout(
    barmode='group',
    title={'text': 'Mental Health Symptom Prevalence by Gender Orientation', 'x': 0.5, 'xanchor': 'center', 'yanchor': 'top', 'font': {'size': 18}},
    xaxis_title={'text': 'Gender Identity', 'font': {'size': 14}},
    yaxis_title={'text': 'Prevalence (% of Population)', 'font': {'size': 14}},
    legend_title_text='Statistics',
    height=500
)

fig.show()
```

```python
df3=pd.read_excel('CIS 591_ Indicators of Anxiety and Depression Data.xlsx', sheet_name='Age')

age_groups = df3[df3['Subgroup'].str.contains('years')]
age_group_description = age_groups.groupby('Subgroup')['Value'].describe()
age_group_description
```

```python
import pandas as pd
import plotly.graph_objects as go

df3 = pd.read_excel('CIS 591_ Indicators of Anxiety and Depression Data.xlsx', sheet_name='Age')

age_groups = df3.loc[df3['Subgroup'].str.contains('years')]

age_groups.loc[:, 'Time Period Start Date'] = pd.to_datetime(age_groups['Time Period Start Date'])

age_groups = age_groups.sort_values(by='Time Period Start Date')

age_groups['Month_Year'] = age_groups['Time Period Start Date'].dt.strftime('%B %Y')

time_grouped = age_groups.groupby(['Subgroup', 'Time Period Start Date'])['Value'].mean().reset_index()

fig = go.Figure()

for age_group in time_grouped['Subgroup'].unique():
    filtered_data = time_grouped[time_grouped['Subgroup'] == age_group]

    fig.add_trace(go.Scatter(
        x=filtered_data['Time Period Start Date'],
        y=filtered_data['Value'],
        mode='lines+markers',
        name=age_group
    ))

fig.update_layout(
    title={'text': 'Trends in Mental Health Symptoms by Age Group', 'x': 0.5, 'xanchor': 'center', 'yanchor': 'top', 'font': {'size': 18}},
    xaxis_title={'text': 'Time Period', 'font': {'size': 14}},
    yaxis_title={'text': 'Prevalence (% of Total Population)', 'font': {'size': 14}},
    xaxis_tickformat='%B %Y',
    xaxis_tickangle=-45,
    legend_title_text='Age Group',
    height=600
)

fig.show()
```

### **Processing Data to CSV File**

```python
gender_stats = gender_stats.reset_index(drop=True)
age_group_description = age_group_description.reset_index()

with pd.ExcelWriter('CIS591_Team14_TS3_Data.xlsx') as writer:
    pivot_table.to_excel(writer, sheet_name='Symptoms by State', index=False)
    top_10.to_excel(writer, sheet_name='Top 10 States', index=False)
    arizona_data.to_excel(writer, sheet_name='Symptoms in AZ', index=False)
    average_values_styled.to_excel(writer, sheet_name='Avg Values in USA', index=False)
    gender_stats.to_excel(writer, sheet_name='Symptoms by Gender', index=False)
    age_group_description.to_excel(writer, sheet_name='Symptoms by Age Group', index=False)


print("Processed data has been saved to 'CIS591_Team14_TS3_Data' with multiple sheets.")
```
